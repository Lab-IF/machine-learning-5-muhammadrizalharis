## A. NumPy (20%)

Tugas:
1. Buat array random 20 angka (0–100)
2. Hitung mean, median, std, min, max
3. Ubah ke matriks 4x5
4. Tampilkan shape dan transpose

```python
arr = np.random.rand(20) * 100

print("Array (20 angka):")
print(arr.round(2))

mean_val = arr.mean()
median_val = np.median(arr)
std_val = arr.std()
min_val = arr.min()
max_val = arr.max()

print("\nStatistik:")
print("Mean  :", round(mean_val, 2))
print("Median:", round(median_val, 2))
print("Std   :", round(std_val, 2))
print("Min   :", round(min_val, 2))
print("Max   :", round(max_val, 2))

mat = arr.reshape(4, 5)

print("\nMatriks 4x5:")
print(mat.round(2))
print("\nShape matriks:", mat.shape)
print("\nTranspose:")
print(mat.T.round(2))
```

##  PENJELASAN BAGAIN A
Pada tahap ini dibuat 20 bilangan acak menggunakan np.random.rand(20). Nilai awal berada pada rentang 0–1, lalu dikalikan 100 sehingga berada pada interval 0–100. Hasilnya dibulatkan hingga dua desimal menggunakan round(2) agar lebih mudah dibaca.

Selanjutnya dihitung statistik dasar meliputi mean, median, standar deviasi, nilai minimum, dan maksimum untuk memperoleh gambaran distribusi data.

Array satu dimensi kemudian diubah menjadi matriks berukuran 4×5 melalui reshape(4,5) guna memahami representasi data dalam bentuk tabel. Terakhir, digunakan shape untuk melihat dimensi matriks dan .T (transpose) untuk menukar baris dan kolom, yang merupakan konsep penting dalam pengolahan data berbasis matriks pada machine learning.

## B. Pandas (25%)

Tugas:
1. Buat DataFrame 10 mahasiswa (Nama, Nilai)
2. Tambahkan kolom Status: Lulus jika nilai ≥ 75, selain itu Tidak Lulus
3. Tampilkan:
   - 3 nilai tertinggi
   - Rata-rata nilai kelas
   - Jumlah mahasiswa lulus

```python
nama = [f"Mahasiswa_{i+1}" for i in range(10)]
nilai = np.random.randint(50, 101, size=10)  # 50–100

df_mhs = pd.DataFrame({
    "Nama": nama,
    "Nilai": nilai
})

df_mhs["Status"] = np.where(df_mhs["Nilai"] >= 75, "Lulus", "Tidak Lulus")

print("DataFrame Mahasiswa:")
df_mhs
```

## Penjelasan Bagian (DATA FRAME MAHASISAAW)

Pada bagian ini dibuat daftar nama mahasiswa secara otomatis dari 1 hingga 10 dengan format Mahasiswa_1 sampai Mahasiswa_10 menggunakan perulangan, sehingga tidak perlu diketik manual.

Nilai mahasiswa dihasilkan secara acak menggunakan np.random.randint(50, 101, size=10), yang menghasilkan 10 angka dalam rentang 50–100.

Data nama dan nilai kemudian disusun ke dalam DataFrame melalui pd.DataFrame(), membentuk tabel dengan kolom Nama dan Nilai.

Selanjutnya ditambahkan kolom Status menggunakan np.where(), di mana nilai ≥ 75 diberi keterangan Lulus, sedangkan nilai < 75 diberi Tidak Lulus.

Terakhir, DataFrame ditampilkan untuk melihat keseluruhan data mahasiswa beserta nilai dan status kelulusannya.

```python
top3 = df_mhs.nlargest(3, "Nilai")
print("Top 3 Nilai:")
display(top3)

rata2 = df_mhs["Nilai"].mean()
print("\nRata-rata nilai kelas:", round(rata2, 2))

jml_lulus = (df_mhs["Status"] == "Lulus").sum()
print("Jumlah mahasiswa lulus:", int(jml_lulus))
```

## Penjelasan Bagian (Mencari Nilai Tertinggi Mahasiswa)

- df_mhs.nlargest(3, "Nilai") digunakan untuk mengambil tiga mahasiswa dengan nilai paling tinggi. Fungsi nlargest() akan mengurutkan data berdasarkan kolom Nilai secara menurun, lalu memilih jumlah data yang ditentukan.

- display(top3) berfungsi menampilkan hasil tiga nilai tertinggi dalam format tabel agar lebih terstruktur dan mudah dibaca.

- df_mhs["Nilai"].mean() digunakan untuk menghitung rata-rata nilai seluruh mahasiswa sebagai indikator performa kelas secara umum.

- round(rata2, 2) membulatkan hasil rata-rata hingga dua angka desimal agar tampilannya lebih rapi.

- (df_mhs["Status"] == "Lulus").sum() menghitung jumlah mahasiswa yang berstatus Lulus dengan menjumlahkan nilai boolean yang sesuai.

- int(jml_lulus) memastikan hasil akhir ditampilkan sebagai bilangan bulat tanpa angka desimal.

## C. Visualisasi (20%)

Buat 2 grafik dari data nilai:
1. Bar chart nilai mahasiswa
2. Histogram distribusi nilai

Syarat:
- Ada judul
- Ada label sumbu
- Ada grid
- Simpan sebagai PNG

```python
plt.figure(figsize=(9,4))
plt.bar(df_mhs["Nama"], df_mhs["Nilai"])
plt.title("Bar Chart Nilai Mahasiswa")
plt.xlabel("Nama Mahasiswa")
plt.ylabel("Nilai")
plt.grid(True, axis="y", linestyle="--", alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("pertemuan01_bar.png")
plt.show()

print("✅ Bar chart disimpan: pertemuan01_bar.png")
```

## Visualisasi (Bar Chart)

- plt.title() menambahkan judul grafik agar isi visualisasi mudah dipahami.

- plt.xlabel() dan plt.ylabel() memberi keterangan pada sumbu horizontal (nama mahasiswa) dan vertikal (nilai).

- plt.tight_layout() menyesuaikan tata letak supaya elemen grafik tertata rapi.

- plt.savefig() menyimpan grafik dalam format PNG, sedangkan plt.show() menampilkan grafik di layar.

- print() memberikan notifikasi bahwa file grafik telah berhasil disimpan.

```python
plt.figure(figsize=(6,4))
plt.hist(df_mhs["Nilai"], bins=5, edgecolor="black")
plt.title("Histogram Distribusi Nilai")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.savefig("pertemuan01_hist.png")
plt.show()

print("✅ Histogram disimpan: pertemuan01_hist.png")
```

- plt.ylabel("Frekuensi") memberi label pada sumbu vertikal yang menunjukkan jumlah mahasiswa dalam setiap rentang.

- plt.grid(True, linestyle="--", alpha=0.5) menambahkan garis bantu agar perbedaan tinggi batang lebih mudah diamati.
